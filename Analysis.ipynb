```python
import pandas as pd
```

```python
# Load the data
historical_data_df = pd.read_csv('historical_data.csv')
fear_greed_index_df = pd.read_csv('fear_greed_index.csv')
```

```python
# Document rows and columns
print("historical_data shape:", historical_data_df.shape)
print("fear_greed_index shape:", fear_greed_index_df.shape)
```

```python
# Document NULLs
# .isnull().sum() counts how many empty/null cells exist in each column of the dataset
print("\n--- Missing Values in historical_data ---")
print(historical_data_df.isnull().sum())

print("\n--- Missing Values in fear_greed_index ---")
print(fear_greed_index_df.isnull().sum())
```

```python
# Document Duplicates
# .duplicated().sum() counts how many completely identical rows exist in the dataset
historical_data_duplicates = historical_data_df.duplicated().sum()
print(f"\nNumber of duplicate rows in Trades: {historical_data_duplicates}")

fear_greed_index_duplicates = fear_greed_index_df.duplicated().sum()
print(f"\nNumber of duplicate rows in Trades: {fear_greed_index_duplicates}")
```

```python
#Displaying the first 10 records
print("\n--- Historical_Data ---")
display(historical_data_df.head(10))

print("\n--- Fear_Greed_Index ---")
display(fear_greed_index_df.head(10))
```

```python
# Datatypes of the datasets
print("\n--- Historical_Data ---")
print(historical_data_df.dtypes)

print("\n--- Fear_Greed_Index ---")
print(fear_greed_index_df.dtypes)
```

```python
# Historical Dataset
# Create a temporary column that holds the full date AND time object
historical_data_df['full_datetime'] = pd.to_datetime(historical_data_df['Timestamp IST'], format='mixed', dayfirst=True)

# Look at that temporary column, extract just the date, and put it in your final column
historical_data_df['date'] = historical_data_df['full_datetime'].dt.date

print("--- Historical_Data with Extracted Date ---")
display(historical_data_df[['Timestamp IST', 'date']].head())
```

```python
#Fear_Greed_Index Data set
#Covertinf the datatype from oject to date
fear_greed_index_df['date'] = pd.to_datetime(fear_greed_index_df['date']).dt.date

print("\n--- Fear_Greed_Index Date ---")
display(fear_greed_index_df[['date']].head())
```

```python
#Merge the datesets on Date
merged_df = pd.merge(historical_data_df, fear_greed_index_df, on='date', how='inner')

print("\n--- Merged Dataset ---")
display(merged_df.head())
```

```python
# Calculate if each individual trade was a win (Closed PnL greater than 0)
# This returns True (1) or False (0) for every row
merged_df['is_win'] = merged_df['Closed PnL'] > 0
display(merged_df[['is_win']].head(100))
```

```python
# Create and diplay an indicator for Long vs Short directions to compute the ratio later
merged_df['is_long'] = merged_df['Direction'].str.lower().str.contains('long').astype(int)
print("\n--- Long ---")
display(merged_df[['is_long']].head(100))

merged_df['is_short'] = merged_df['Direction'].str.lower().str.contains('short').astype(int)
print("\n--- Short ---")
display(merged_df[['is_short']].head(100))
```

```python
# Group the data by Date and Account to calculate daily summaries (Excel PivotTable equivalent)
daily_metrics = merged_df.groupby(['date', 'Account']).agg(
    daily_pnl=('Closed PnL', 'sum'),             # Total profit or loss made per day
    num_trades=('Trade ID', 'count'),            # Total number of trades executed per day
    avg_trade_size=('Size USD', 'mean'),         # Proxy for risk/leverage (Average size in USD)
    total_wins=('is_win', 'sum'),                 # Count how many trades won
    total_longs=('is_long', 'sum'),               # Count how many long trades were made
    total_shorts=('is_short', 'sum')              # Count how many short trades were made
).reset_index()

# Display a preview of our newly created daily matrix sheet
print(daily_metrics.head(100))
```

```python
# Calculate the Win Rate percentage (Wins divided by Total Trades)
daily_metrics['win_rate'] = daily_metrics['total_wins'] / daily_metrics['num_trades']
display(daily_metrics[['win_rate']].head(100))
```

```python
# Calculate the Long/Short Ratio (Longs divided by Shorts, handling division by zero with max)
daily_metrics['long_short_ratio'] = daily_metrics['total_longs'] / daily_metrics['total_shorts'].replace(0, 1)
display(daily_metrics[['long_short_ratio']].head(100))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
# Group by the Fear & Greed classification to compare baseline trading performance metrics
sentiment_analysis = merged_df.groupby('classification').agg(
    avg_pnl_per_trade=('Closed PnL', 'mean'),     # Average money made/lost per trade
    overall_win_rate=('is_win', 'mean'),          # Overall percentage of winning trades
    avg_position_size=('Size USD', 'mean'),       # Average volume allocated per position
    total_trade_count=('Trade ID', 'count')       # Total transaction frequency
).reset_index()

# Display the performance overview table
print(sentiment_analysis)
```

```python
# Average PnL Bar Chart

# Create color dictionary
sentiment_colors = {
    'Extreme Fear': '#cc0000',  # Dark Red
    'Fear': '#ff9999',          # Light Red
    'Neutral': '#cccccc',       # Grey
    'Greed': '#99ff99',         # Light Green
    'Extreme Greed': '#009900'  # Dark Green
}

# Start a fresh, standalone figure
plt.figure(figsize=(8, 6))

# Build the Bar Chart with the expanded palette and hue
sns.barplot(
    data=sentiment_analysis, 
    x='classification', 
    y='avg_pnl_per_trade', 
    hue='classification', 
    palette=sentiment_colors,
    legend=False # Turns off the redundant legend since the x-axis is already labeled
)

# Formatting directly on plt
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Average PnL per Trade', fontsize=14, fontweight='bold')
plt.xlabel('Market Sentiment', fontsize=12)
plt.ylabel('Average PnL (USD)', fontsize=12)

# Display just this chart
plt.show()
```

```python
#Chart for Win Rate
# Start a second fresh, standalone figure
plt.figure(figsize=(8, 6))

# Build the second Bar Chart using the exact same color dictionary
sns.barplot(
    data=sentiment_analysis, 
    x='classification', 
    y='overall_win_rate', 
    hue='classification',
    palette=sentiment_colors,
    legend=False 
)

# Formatting directly on plt
plt.axhline(0.5, color='black', linestyle='--', linewidth=1.5, label='50% Breakeven')
plt.title('Overall Win Rate', fontsize=14, fontweight='bold')
plt.xlabel('Market Sentiment', fontsize=12)
plt.ylabel('Win Rate', fontsize=12)

# Display just this chart
plt.show()
```

```python
#Trade Frequency Chart
# Create color dictionary
sentiment_colors = {
    'Extreme Fear': '#cc0000',  # Dark Red
    'Fear': '#ff9999',          # Light Red
    'Neutral': '#cccccc',       # Grey
    'Greed': '#99ff99',         # Light Green
    'Extreme Greed': '#009900'  # Dark Green
}

# Generate Trade Frequency figure
plt.figure(figsize=(8, 6))
sns.barplot(
    data=sentiment_analysis, 
    x='classification', 
    y='total_trade_count', 
    hue='classification',
    palette=sentiment_colors,
    legend=False
)

# Apply formatting directly to this figure
plt.title('Trade Frequency by Sentiment', fontsize=14, fontweight='bold')
plt.xlabel('Market Sentiment', fontsize=12)
plt.ylabel('Total Number of Trades', fontsize=12)

# Display just this chart
plt.show()
```
